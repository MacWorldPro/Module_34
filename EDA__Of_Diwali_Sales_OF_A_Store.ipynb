```python
import numpy as np
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
```

```python
df=pd.read_csv('data.csv', encoding= 'unicode_escape')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

Cleaning the data

```python
df.columns
```

```python
df.drop(columns=['Status','unnamed1'],axis=1,inplace=True)
```

```python
# Checking out for the outliers
df['Amount'].hist()
plt.show()
```

```python
mean=np.mean(df['Amount'])
df['Amount'].fillna(mean,inplace=True)
```

```python
df['Amount']=df['Amount'].astype('int')
```

Starting the process of EDA

Checking for different age-groups who did shopping and also their shopping amount

```python
fig=sns.countplot(data=df,x='Age Group',hue='Gender')
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

State Wise

```python
fig = go.Figure()
fig.add_trace(go.Histogram(x=df['State']))
fig.show()
```

```python
zones = df['Zone'].unique()
fig, axes = plt.subplots(1, len(zones), figsize=(12, 6), sharey=True)
for i, zone in enumerate(zones):
    sns.countplot(data=df[df['Zone'] == zone], x='State', hue='Gender', ax=axes[i])
    axes[i].set_title(f'Zone {zone}')

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

```python
# Total Amount vs State
sales_age = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'State',y= 'Amount' ,data = sales_age)
```

So by sate wise we can coclude that state of uttarpradesh has got the total sales and that too in the central zone, most candiates are female.

Let's do analysis occupation  wise.

```python
fig = go.Figure()
fig.add_trace(go.Histogram(x=df['Occupation']))
fig.show()
```

```python
fig=sns.countplot(data=df,x='Occupation',hue='Gender')
```

```python
# Total Amount vs Occupation
sales_age = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Occupation',y= 'Amount' ,data = sales_age)
```

It sector has got the most number of sales and that too from women and in terms of total amount also.
