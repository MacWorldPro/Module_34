```python
# extracting data for particular players
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Step 1: Scrape the player data from the Wikipedia page
url = 'https://en.wikipedia.org/wiki/2022_Indian_Premier_League_final'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Find the section with the player names
# We will use a CSS selector to target the player tables
tables = soup.select('table')

# Extract player names from the two teams
player_names = []
for table in tables:
    for row in table.find_all('tr'):
        cells = row.find_all('td')
        if len(cells) > 0:
            player_name = cells[0].get_text(strip=True)
            player_names.append(player_name)

# Print the player names for verification
print(player_names)

# Step 2: Filter the dataset
# Assuming df_combined is your DataFrame with player data
# Ensure that player names are cleaned and standardized in both the DataFrame and the scraped list

# df_combined['fullName'] = df_combined['fullName'].str.strip()  # Clean whitespace

# # Filter the DataFrame to only include rows where 'fullName' is in the scraped player_names list
# filtered_df = df_combined[df_combined['fullName'].isin(player_names)]

# # Print the filtered DataFrame
# print(filtered_df)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

def create_features(dataframe):
    player_models = {}
    for name, group in dataframe.groupby('fullName'):
        if len(group) >= 5:  # Ensure at least 5 data points for prediction
            X = []
            y = {col: [] for col in group.columns[1:]}

            for i in range(2, len(group)):
                feature_row = []
                for col in group.columns[1:]:
                    feature_row.extend(group[col].iloc[i-2:i+1].values)  # Take last 3 matches including current match
                    y[col].append(group[col].iloc[i])  # Target is the next match's value for each feature
                X.append(feature_row)

            # Ensure X and y have consistent lengths
            X = np.array(X)
            for col in y:
                y[col] = np.array(y[col])
                if len(X) != len(y[col]):
                    raise ValueError(f"Inconsistent number of samples for {col}: X has {len(X)} samples, y has {len(y[col])} samples.")

            player_models[name] = {'X': X, 'y': y}

    return player_models

# Prepare data
 # Replace with actual data loading step
player_models = create_features(filtered_df_copy)

# Train models for each eligible player and each feature
for player, data in player_models.items():
    X = data['X']
    y = data['y']

    # Train models for each feature
    models = {}
    for feature, target_values in y.items():
        if len(X) > 1:  # Ensure there's enough data to split
            # Split data into train and test sets
            X_train, X_test, y_train, y_test = train_test_split(X, target_values, test_size=0.2, random_state=42)

            model = LinearRegression()
            model.fit(X_train, y_train)
            models[feature] = model

    player_models[player]['models'] = models

# Predicting future performance
predictions = {}
for player, data in player_models.items():
    if len(data['X']) >= 1 and 'models' in data:  # Ensure there is at least one match to predict and models exist
        models = data['models']
        X_last_3_matches = data['X'][-1]  # Last 3 matches' data

        player_predictions = {}
        for feature, model in models.items():
            prediction = model.predict([X_last_3_matches.flatten()])[0]
            player_predictions[feature] = prediction

        predictions[player] = player_predictions

# Convert predictions to DataFrame
predictions_df = pd.DataFrame(predictions).T


```
