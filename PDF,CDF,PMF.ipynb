```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

PLotting PMF + CDF of Discrete Values

```python
# Compute PMF
pmf = df['size'].value_counts(normalize=True).sort_index()

# Plot PMF with curve-like structure
sns.kdeplot(df['size'], shade=True)
# plt.plot(pmf.index, pmf.values, marker='o', linestyle='-')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Probability Mass Function (PMF) - Discrete Values')
plt.show()
```

```python
# Compute CDF
cdf = pmf.cumsum()

# Plot CDF
plt.step(cdf.index, cdf.values, where='mid')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function (CDF) - Discrete Values')
plt.show()
```

Computing PDF + CDF of continuous Values

```python
# Compute PDF
pdf = df['total_bill'].value_counts(normalize=True).sort_index()

sns.kdeplot(df['total_bill'], shade=True)
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Probability Density Function (PDF)')
plt.legend()
plt.show()
```

```python
# Calculate z-scores
z_scores = (df['total_bill'] - np.mean(df['total_bill'])) / np.std(df['total_bill'])
# Plotting standard normal distribution
sns.kdeplot(z_scores, shade=True)
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Standard normal distribution')
plt.legend()
plt.show()
```

```python
# Compute CDF
cdf = np.cumsum(pdf)
# Plot CDF
plt.step(cdf.index, cdf.values, where='mid', label='CDF')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function (CDF)')
plt.legend()
plt.show()
```

```python
# PDF,PMF,CDF are representations
# of continuous,discrete and cumulative values.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# Bernoulli distribution parameters
p = 0.5  # Probability of success

# Number of samples to draw
num_samples = 1000

# Number of trials in each sample
sample_size = [1, 2, 5, 10, 50, 100]  # Sample sizes to demonstrate CLT

# Plotting the Bernoulli distribution
plt.figure(figsize=(12, 8))
plt.subplot(3, 3, 1)
plt.title("Bernoulli Distribution")
plt.bar([0, 1], [1 - p, p], tick_label=['0', '1'])
plt.xlabel("Outcome")
plt.ylabel("Probability")

# Sampling and plotting the sample means
for i, n in enumerate(sample_size, start=2):
    sample_means = []
    for _ in range(num_samples):
        sample = bernoulli.rvs(p, size=n)  # Generating a sample of size n
        sample_mean = np.mean(sample)      # Calculating the sample mean
        sample_means.append(sample_mean)

    plt.subplot(3, 3, i)
    plt.hist(sample_means, bins=30, density=True, alpha=0.6)
    plt.title(f"Sample Mean Distribution (n={n})")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")

plt.tight_layout()
plt.show()
```
