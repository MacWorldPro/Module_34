```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom

# Parameters
p = 0.3  # Probability of success for both Bernoulli and Binomial distributions
n = 10   # Number of trials for the Binomial distribution

# Generate data points for Bernoulli distribution
data_bernoulli = bernoulli.rvs(p, size=1000)

# Generate data points for Binomial distribution
data_binomial = binom.rvs(n, p, size=1000)

# Plotting the distributions
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Bernoulli distribution
axs[0].hist(data_bernoulli, bins=[-0.5, 0.5, 1.5], density=True, alpha=0.6)
axs[0].set_title('Bernoulli Distribution')
axs[0].set_xticks([0, 1])
axs[0].set_xticklabels(['Failure', 'Success'])
axs[0].set_ylabel('Probability')

# Binomial distribution
axs[1].hist(data_binomial, bins=np.arange(-0.5, n+1.5, 1), density=True, alpha=0.6)
axs[1].set_title('Binomial Distribution (n=10)')
axs[1].set_xlabel('Number of Successes')
axs[1].set_ylabel('Probability')

plt.tight_layout()
plt.show()
```
